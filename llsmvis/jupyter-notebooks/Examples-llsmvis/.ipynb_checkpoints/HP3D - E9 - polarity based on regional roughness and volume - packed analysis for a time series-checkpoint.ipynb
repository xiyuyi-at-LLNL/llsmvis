# Write out the polarity characterization output based on patch-wise volume and roughnesses

```python
pcs=[]
ncs=[]
pstds=[]
nstds=[]
pdvs=[]
ndvs=[]
polarvs=[]
for timeind in np.arange(50):
# for timeind in [0]:
    print(timeind)
    # load the data
    fpath = '2021-11-17-cell10'
    conespath = '/p/lscratchh/yi10/hp3d/'+fpath+'/cone_cuts_analysis/time'+str(timeind).zfill(4)
    h5path = os.path.join(conespath, 'patch-wise-roughness-volumes.hdf5')
    f = h5py.File(h5path, 'a')
    rou = np.asarray(f['roughnesses on cone vectors'])
    vol = np.asarray(f['cone volumes on cone vectors'])
    roumap = np.asarray(f['roughnesses on map'])
    volmap = np.asarray(f['cone volumes on map'])

    # get the cone angles
    sample_angle = 15  # sampling angle when choose semi-equal spaced points.
    xs,ys,zs,thetalist,philist = v.sample_points_semi_equal_space(sample_angle, output_angles=True)
    cone_vectors=np.asarray([xs, ys, zs])
    

    # get the v centers
    [pc, pstd, pdv, nc, nstd, ndv, polarity_vector] = \
                    sm.get_pos_neg_vcenters(v = vol, cone_vectors = [xs, ys, zs],normalize_v=True)
    
    # get the roughness centeres on the map
    maximavalues, maximatags, maximainds, maximathes, \
    maximaphis, minimavalues, minimatags, minimainds, minimathes, minimaphis = \
    sm.mark_local_extrema_on_maps(cone_vectors, thetalist, philist, 
                                  roughnesses=rou, roumap=roumap, neighbour_n=12,
                                 show_plot=False)
    
    # write out to the hdf5 file
    tools.polarity_vr_writeout(f, pc,pstd,pdv,nc,nstd,ndv,vol,polarity_vector,\
                         maximavalues, maximatags, maximainds, maximathes, maximaphis, \
                         minimavalues, minimatags, minimainds, minimathes, minimaphis )
    
    for a in f.keys():
        print(a)
    f.close()
    pcs.append(pc)
    ncs.append(nc)
    pstds.append(pstd)
    nstds.append(nstd)
    pdvs.append(pdv)
    ndvs.append(ndv)
    polarvs.append(polarity_vector)
    
    
```
