```python
def ck(timeind):
    fpath='2021-11-17-cell9'
    conespath='/p/lscratchh/yi10/hp3d/'+fpath+'/cone_cuts_analysis/time'+str(timeind).zfill(4)
    h5path=os.path.join(conespath, 'patch-wise-roughness-volumes.hdf5')
    f = h5py.File(h5path, 'r')
    f.keys()
    a=np.asarray(f['roughnesses on map'])
    b=np.asarray(f['cone volumes on map'])
    plt.figure(figsize=(15,15))
    plt.subplot(1,2,1)
    plt.imshow(a)
    plt.subplot(1,2,2)
    plt.imshow(b)
    f.close()
```
