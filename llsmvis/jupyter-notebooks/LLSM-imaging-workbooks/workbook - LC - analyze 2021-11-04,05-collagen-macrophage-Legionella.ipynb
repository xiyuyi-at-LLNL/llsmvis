```python
# define result folder path and Positions.csv file path
raw_data_folder = '/p/lscratchh/yi10/llsm-data/2021-11-02-find-engulfed-Legionella/Fila_C0_tlaps/ch0'
result_folder = raw_data_folder+'/results_dsk'


make_new_csv = True
import sys
sys.path.append("/g/g92/yi10")
from llsmvis import *
from llsmvis.parser import *
from llsmvis import utils
import skvideo.io
import warnings; warnings.simplefilter('ignore')
# import stuff...
import os
from skimage import io
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
csubs='Filamentous_C0_tlaps'
VERBOSE=True
gp=parser.LLSMGroupParser1s1t(raw_data_folder,csubs) # define the group parser
gp.prefer_deskewed=False # this will ignore all the deskewed tiffs
gp.get_indiv_parsers() # this is just looping over each individual tiff stack and generate an individual parser using the older single paser.
print('validity is '+str(gp.valid_check()))
gp.get_parser_time_indexes() 
gp.sort_parsers_by_time() # this sorts the parsers in the group by time.
```

```python
tlist=gp.time_sorted_indiv_parsers  # a list of parsers with time order
```

```python
lb=0
ub=2000
ptlist=gp.time_sorted_indiv_parsers  # a list of parsers with time order
t0=datetime.strptime(ptlist[0].acquisition_startT,'%m/%d/%Y %I:%M:%S %p')
channel_inds=None
if channel_inds is None:
    channel_inds=np.arange(0,len(ptlist[0].channel_names))
cind=0
c=ptlist[0].channel_names[cind]
c = ptlist[0].channel_names[cind]
# find the correct file name for the MIP project in XY of this channel.
fname0 = ptlist[0].fpath+'/results_dsk/'+ptlist[0].fname_head + '_deskewed'+'/MIP_'+c+'_XY.tif'
imstack = skimage.external.tifffile.imread(fname0).astype('float')  
peak = np.max(imstack.ravel());
base = np.min(imstack.ravel());
imstack = (imstack-base)/(peak-base)
plt.figure(figsize=(15,15))
plt.imshow(imstack,vmin=lb, vmax=ub,cmap='gray')
print(ub)
```
