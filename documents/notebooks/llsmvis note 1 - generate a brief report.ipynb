# llsmvis note 1 - Generate a brief report of the data

This note shows how to generate a brief report of a dataset acquired from the lattice light sheet microscope with multiple channels and multiple time steps using <span style="color:blue">llsmvis</span>.

#### First, import necessary tools:

```python
import sys
sys.path.append('/Users/yi10/Desktop/Research/Software/T-llsmvis') # replace this to the full path of your llsmvis folder.
from llsmvis import *
from llsmvis import utils
import skvideo.io
import warnings; warnings.simplefilter('ignore')
```

#### Second, define the datapath, and the file name head as follows, and parse the data using the parser

```python
# define fpath and name_head
fpath='/Volumes/SSD1/LLSM_datasets/20200825-tweak'
fname_head = 'cell2_2'

# parse the data
p = LLSMParser(fpath, fname_head)
# specify channel names
p.set_channel_names(['mCherry']) # this will overwrite the default channel_names from ['channel0','channel1', etc...]
```

```python
p.all_tiffs
```

#### Third, create the table with basic data properties, and display as a picture

```python
[fig, ax] = utils.render_mpl_table(p, p.data_properties, header_columns=0, col_width=7.0)
```

You can drag the picture above into your .ppt file for your presentation. You can also right click on the picture to save the image.
